
A mutual fund is a professionally managed investment fund that pools money from many investors to purchase securities. These investors may be retail or institutional in nature. 



Mutual funds are operated by companies known as Asset Management Companies (AMC).

For example, Aditya Birla is an AMC that operates several different mutual fund schemes like Aditya Birla Sun Life etc.

Risk of a mutual fund depends on which mutual fund we invest in.
Mutual funds can give returns from 4% to 30%.


# Types of Mutual Funds:

Mutual Funds are basically classified into 3 catagories:

1) Equity Funds: Mutual Funds that invest on stock market. These funds ususally have high risk and high return.Equity funds are commonly classified into several types.

a) Large Cap Equity Funds: Mutual Funds which invest in stocks of big companies. It  has less risk but low growth

b) Small/Mid Cap equity funds: Mutual funds which invest in stocks of small and medium sized companies. There is high risk as      high growth in these funds.

c) Diversified Equity Funds: Also known as multi-cap fund because investment is done on large cap as well as small and mid-cap   too.

d)Tax saving equity fund: Equity funds where one can tax worth upto Rs.1.5 lakh.

e)Sector Mutual Funds: Invested specifically on big sectors such as agriculture,logistics,food processing etc. These funds are usually very risky.

f)Index Funds: Passively managed funds.They are dependent on Sensex and Nifty.

2) Debt Funds:  Are a collection of fixed income securities and thus are diversified in nature. The investment can be done through government securities,corporate bonds, commercial papers etc. Debt funds are usually safe as give promise returns but the returns are usully low.
 Debt funds are commonly classified into several types:
 
 a) Liquid Fund: These funds that can be easily converted into cash. These funds have low risk.
 
 b) Gilt Fund: When government issues bonds with zero risk with fluctating interest rates is known as Gilt Funds.
 
 c) Fixed maturity plan: It is an alternative to Fixed Deposit. It depends on the time of investment.
 
 d) Money Market Funds:A money market fund (also called a money market mutual fund) is an open-ended mutual fund that invests      in short-term debt securities.
 
3) Hybrid Funds: Mutual Funds that invest both on stock market and on bonds.Hybrid funds are commonly classified into several       types:

a) Balanced Saving Fund (Equity Savings): These funds have majority of debt funds and minority of equity funds.
   Nominal Ratio [ 70 (Debt): 30 (Equity) ]
   
b)Balanced Adventage Fund (Hybrid Agressive):These funds have majority of equity funds and minority of debt funds.
   Nominal Ratio [ 70 (Equity) : 30 (Debt) ]

The 20 different parameters are-

1)Sl_No.: Serial no. of the schemes of mutual funds. They are ranked upon 1 year returns in descending order.

2)Fund_Scheme_Name: Name of the scheme.

3)Launch_Date: The date on which the scheme has begun.

4)Net_Assets_(Cr): Assets under management (AUM), sometimes called funds under management (FUM), measures the total market value of all the financial assets which a financial institution such as a mutual fund, venture capital firm, or broker manages on behalf of its clients and themselves.

5)1_Yr_Return- How much return is gained after investing for one year in that scheme.

9)Crisil_Rating-CRISIL( formely Credit Rating Information Services of India Ltd.) is a global analytic company  providing ratings, research,risk and policy advisory. An AMC with a  Crisil Rating of 5 is usually said to be the best performing AMC.

10)Risk: Risk in a mutual fund states states that you could lose your money on your investment at that particular AMC.

11)Alpha:Alpha measures the difference between the fund's actual returns and its expected performance,given its level of risk which is measured as Beta. A positive Alpha indicates that the fund has performed better than the Beta would predict. A negative Alpha indicates that the fund has underperformed given the expectations established by funds's Beta.

12)Beta:Beta measures the responsiveness of a stock's price to changes in the overall stock market. On comparison of the benchmark index for e.g. NSE Nifty to a particular stock returns, a pattern develops that shows the stock's openness to the market risk. This helps the investor to decide whether he wants to go for the riskier stock that is highly correlated with the market (beta above 1), or with a less volatile one (beta below 1).

For example, if a stock's beta value is 1.3, it means, theoretically this stock is 30% more volatile than the market. Beta calculation is done by regression analysis which shows security's response with that of the market.

By multiplying the beta value of a stock with the expected movement of an index, the expected change in the value of the stock can be determined. For example, if beta is 1.3 and the market is expected to move up by 10%, then the stock should move up by 13% (1.3 x 10).

13)R_Squared:R-squared measures the relationship between a portfolio and its benchmark. It can be thought of as a percentage from 1 to 100.R-squared is not a measure of the performance of a portfolio. A great portfolio can have a very low R-squared. It is simply a measure of the correlation of the portfolio's returns to the benchmark's returns.R-squared can be used to ascertain the significance of a particular beta or alpha. Generally, a higher R-squared will indicate a more useful beta figure. If the R-squared is lower, then the beta is less relevant to the fund's performance.

14)Std_Dev: Standard deviation (SD) is a measure of volatility of how far the measurement such as return tends to deviate from an average over particular period.

15)Sharpe_Ratio:Sharpe ratio is the measure of risk-adjusted return of a financial portfolio. A portfolio with a higher Sharpe ratio is considered superior relative to its peers.Sharpe ratio is a measure of excess portfolio return over the risk-free rate relative to its standard deviation. Normally, the 90-day Treasury bill rate is taken as the proxy for risk-free rate.

16)Category:A fund category is a way of differentiating mutual funds according to their investment objectives and principal investment features. This categorization allows investors to spread their money around in a mix of funds with a variety of risk and return characteristics.

17)Consitency: Consistency indicates whether the AMC is constantly performing or not by giving constant returns to the depositers.

18)Year Trailing: Indicates how old is the scheme.

19)Sortino_Ratio:Sortino ratio is the statistical tool that measures the performance of the investment relative to the downward deviation. Unlike Sharpe, it doesn't take into account the total volatility in the investment.
Sortino ratio is similar to Sharpe ratio, except while Sharpe ratio uses standard deviation in the denominator, Frank A Sortino uses downside deviation in the denominator.
Standard deviation involves both the upward as well as the downward volatility.

Since investors are only concerned about the downward volatility, Sortino ratio presents a more realistic picture of the downside risk ingrained in the fund or the stock.

20)Net_Returns:It typically includes capital gains and any income received from the investment. By comparison, the net rate of return deducts fees and expenses from the investment's final value. For net return, not only fees and commissions are deducted, but also the effects of taxes and inflation.

# Part3: Verified the new data by deteming its accuracy

Now, the testing and training of data is used by taking 8 parameters that is 'R_Squared','Net_Return','Alpha','Beta','Sharpe_Ratio','Std_Dev','Sortino_Ratio','Expense_Ratio' in the independent variable which can be assumed as X, while the parameter 'Good' is taken in the dependent variable say Y as ultimately our goal is to check which funds are good and which aren't. A test size of 0.2 is assumed for testing and training the data.

```python
#Train test split
from sklearn.model_selection import train_test_split
#Features
X=data[['R_Squared','Net_Return','Alpha','Beta','Sharpe_Ratio','Std_Dev','Sortino_Ratio','Expense_Ratio']]
Y=data['Good'] #Status
#split dataset into into training set and testing set
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)#80% training and 20% testing
```

Accuracy of the data can be calculated using various models. In this we are using AdaBoostClassifier,

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier
#Create a classifer and fit the data
clf=AdaBoostClassifier(random_state=1)
clf.fit(X_train,Y_train)
```

```python
#Find accuracy using test set
Y_pred=clf.predict(X_test)
print('Accuracy:{}'.format(accuracy_score(Y_pred,Y_test)))
```

Accuracy=1.0 states that our data is absolutely accurate

```python
clf.predict([[3,4,5,6,7,8,9,10]])
```

clf.predict helps us in estimating what would be the output if we pass the values according to condition systematically.

# ROC

```python
#Importing roc-auc packages
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
def plot_roc_curve(fpr,tpr):
    plt.plot(fpr,tpr,color='red',label='ROC')
    plt.plot([1,0],[0,1],color='black',linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC(Receiver Operating Curve)')
    plt.legend()
    plt.show()
    
```

```python
probs=clf.predict_proba(X_test)
probs
```

```python
probs=probs[:,1]
probs
```

```python
auc=roc_auc_score(Y_test,probs)
print('AUC:%.2f'%auc)
```

```python
fpr,tpr,thresholds=roc_curve(Y_test,probs)
```

```python
plot_roc_curve(fpr,tpr)
```

AUC-ROC is a performance measurement for classification problem at various threshold settings.ROC is a probability curve and AUC represents the degree of measure of seperability. It tells how much the model is capable to distinguish between the classes. Higher the AUC better the model is at predicting 0s and 0s and 1s and 1s. By analogy, higher the AUC better the model is at predicting the AMC between who is good who is not.

In this case since AUC=1.00 it is very good distinguish a good AMC and bad AMC
